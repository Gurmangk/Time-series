```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hourlysolardata.csv')
```

```python
df
```

```python
df.plot()
```

```python
import seaborn as sns
```

```python
fig, axes = plt.subplots(3, 1, figsize=(11, 10), sharex=True)
for name, ax in zip(['Ambient Temperature (C)', 'DC Array Output (W)', 'AC System Output (W)'], axes):
    sns.boxplot(data=df, x='Hour', y=name, ax=ax)
    ax.set_ylabel('GWh')
    ax.set_title(name)
# Remove the automatic x-axis label from all but the bottom subplot
    if ax != axes[-1]:
        ax.set_xlabel('')
```
